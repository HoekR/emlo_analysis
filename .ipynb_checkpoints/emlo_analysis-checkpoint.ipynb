# Data Scopes at a scale: Analyzing EMLO


For the Data Scopes at a scale article we wanted a dataset to analyze on different scales. A collection of letters seemed a good choice as 

- letters are objects more or less familiar to most humanities scholars 
- letters can be studied at many different scales: a single letter is as much a viable object of study as a large collection of letters and anything in between
    
We opted for the Early Modern Letters Online (http://emlo.bodleian.ox.ac.uk) collection compiled by the Oxford Bodleian Library in cooperation with a host of other institutes. At the time of writing (March 2020) it contains 51 collections covering both the diplomatic and the scientific and cultural world and spans more than three centuries, with the first letter from 1506 and the last from 1837. 

The collection is not available for download, so we had to do some web scraping to get the information ready for analysis.

```python
#aldf.to_csv("/data/data_scopes/all_writers.csv")

aldf = pd.read_csv("/data/data_scopes/all_writers.csv")
```

```python
written_letters = aldf['Letters Written'].str.replace('-','0').astype('int64')
written_letters.value_counts().plot.line()
```

```python
received_letters = aldf['Letters Received'].str.replace('-','0').astype('int64')
received_letters.value_counts().plot.line()
```

```python
mentioned_letters = aldf["Letters Mentioning"].str.replace('-','0').astype('int64')
mentioned_letters.value_counts().plot.line()
```

```python
wl = written_letters.sum()
rl = received_letters.sum()
ml = mentioned_letters.sum()
print(wl, rl, ml)
```

#### Analysis

From this analysis, it appears that there were approximately as many letters received (152506) as letters written (156212). It is clear that there are many less letters mentioning (26349). This is, obviously, because the 'letters mentioning' category requires mining (or reading) letters for names and this has not been done exhaustively.

Let's look into that in some more details

```python
received_letters.sort_values(ascending=False).head(20)
```

```python
written_letters.sort_values(ascending=False).head(20)
```

```python
mentioned_letters.sort_values(ascending=False).head(20)
```

```python
aldf.drop([c for c in aldf.columns if 'Unnamed' in c], axis=1, inplace=True)
aldf["Letters Written"] = written_letters
aldf["Letters Received"] = received_letters
aldf["Letters Mentioning"] = mentioned_letters
aldf
```

```python
aldf.aggregate('sum')
```

```python
# for the distributions see above, but they need some more analysis
# calculated disbalances in the written vs received categories

wisr = aldf.loc[(aldf["Letters Written"] == aldf["Letters Received"])]
wgtr = aldf.loc[(aldf["Letters Written"] > aldf["Letters Received"])]
rgtw = aldf.loc[(aldf["Letters Written"] < aldf["Letters Received"])]
print(len(wisr), len(wgtr), len(rgtw))
```

#### Analysis

Or in words: 

- there is a minority of persons with an equal amount of letters received as written
- there are almost three times as many people who wrote more letters than they received
    

```python
# let's look closer at the rows where the number of received letters == the number of written letters
wisr["Letters Written"].value_counts()
```

```python
#what could those letters with no writer or receiver be?
wisr.loc[wisr["Letters Written"]==0]
```

ah, so this means that there are letters that were probably counted twice. So we would have to compare these numbers with the collections (if that is even possible). We'll look into that later on.

```python
# but what about the other people with equal numbers of written and received letters
wisr.loc[wisr["Letters Written"]!=0].sort_values(by=["Letters Written"], ascending=False)
```

and the number of people with an equal amount of received and written letters is a very small minority. Perhaps their numbers would grow somewhat if the written and received number of letters would not have to be exactly equal, but I see no value in calculating that.
